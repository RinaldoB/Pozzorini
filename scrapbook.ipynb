```python
maxidx = 10000
dt = 1/20000.; time = arange(maxidx)*dt

stim    = fromfile('stim_files/I_noise_test_30s.tpl', dtype='f4')[:maxidx]*1e3
#mV

amp_in  = loadtxt('recordings/Noise_Test_2_20kHz_30s_2016_06_21/Noise_Test_2_CurrentIn_(input)_20kHz_30s_2016_06_21.txt', skiprows=1)[:maxidx]*1e12
#pA

amp_out = loadtxt('recordings/Noise_Test_2_20kHz_30s_2016_06_21/Noise_Test_2_VoltageIn_(output)_20kHz_30s_2016_06_21.txt', skiprows=1)[:maxidx]*1e3
#mV
```

```python
subplot(3,1,1)
plot(time, stim)
subplot(3,1,2)
plot(time, amp_in)
subplot(3,1,3)
plot(time, amp_out)
```
